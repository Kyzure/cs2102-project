# Application Functionalities

Your application must support the following functionalities, each of which is to be implemented as SQL or PL/pgSQL routine (*i.e.,* function or procedure).
Some of these routines return a JSON value.
PostgreSQL provides many useful functions for JSON data type (*e.g.,* [row to JSON](https://www.postgresql.org/docs/current/functions-json.html), [JSON aggregate](https://www.postgresql.org/docs/current/functions-aggregate.html)).

Note that the specification for the routines are *incomplete* by design.
For instance:

* The data types for the input parameters are not specified.
* Certain parameters are not explicitly defined.

For any criterion/process/design issues that are not explicitly stated, you are free to decide on how to address that issue in a reasonable way and justify your decisions in your project report.

1. `add_employee`:
    This routine is used to add a new employee.
    The inputs to this routine includes at least the following:
    * name
    * contact number
    * department
    * kind (this is either junior, senior or manager)
    * manager (may be `NULL`)
    
    The employee ID and email address is automatically generated by the system.
2. `remove_employee`:
    This routine is used to remove an employee.
    The inputs to this routine includes at least the following:
    * employee ID
    
    As before, if the employee is a manager, then we assign all employees supervised by this manager into a dummy manager.
    The employee should also be removed from all future meetings.
